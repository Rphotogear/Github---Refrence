```python
#Handles Dove's report

import numpy as np
import pandas as pd
from datetime import date
from datetime import timedelta


file_products = 'ProductExport-'+date.today().strftime("%d-%m-%Y")+'.csv'
use_col_p = ['Brand',
  'Code',
  'Product Name',
  'Supplier',
  'Supplier Code',
  'Supplier Stock1',
  'Stock Avail',
  'Stock Status Out']

df_p = pd.read_csv(file_products, usecols=use_col_p)
df_p = df_p.fillna("")

def dove():
  df_p_supplier = df_p.loc[df_p['Supplier']=='Dove Electronics Limited']
  df_p_supplier['match'] = df_p_supplier['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  file_path = 'Dove-'+date.today().strftime("%d-%m-%Y")+'.csv'
  use_col = ['Brand', 'Description', 'Dove Code', 'Qty']
  brands_tokeep = ['Adata',
    'AF',
    'Digitus',
    'Ednet',
    'Olympus',
    'Panasonic',
    'Sansai',
    'Sony',
    'Sony Alpha',
    'Synology',
    'Viewsonic',
    'Western Digital'
    ]


  df = pd.read_csv(file_path, usecols=use_col)
  df.rename(columns={'Description': "Name",'Dove Code': "Supplier Code"}, inplace = True)
  df = df.loc[df['Brand'].isin(brands_tokeep)]
  df['Supplier Stock Message'] = "Please contact us to enquire"
  df.loc[df['Qty']>0, 'Supplier Stock Message'] = 'In stock with Supplier'

  df['match'] = df['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  df_supplier_stock = df[['match','Supplier Stock Message']]
  df_p_supplier = df_p_supplier.merge(df_supplier_stock, how='left', on='match').fillna("")

  #Filtered out what needs to be updated

  return df_p_supplier

def lacklands():
  df_p_supplier = df_p.loc[df_p['Supplier']=='Lacklands']
  df_p_supplier['match'] = df_p_supplier['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  file_path = 'Lacklands-'+date.today().strftime("%d-%m-%Y")+'.csv'
  use_col = ['STOCKCODE', 'NAME', 'BRAND', 'ONHAND']
  brands_tokeep = ['AVENGER',
    'CAMELION',
    'ECOFLOW',
    'GITZO',
    'HAHNEL',
    'HENZO',
    'INCA',
    'JOBY',
    'KONUS',
    'LENSPEN',
    'LOWEPRO',
    'MANFROTTO',
    'MARUMI',
    'NIKON',
    'NONDA',
    'PEAK DESIGN',
    'SAMYANG',
    'SAVAGE',
    'SYRP',
    'TAMRON',
    'TTARTISAN',
    'VARTA']

  df = pd.read_csv(file_path, usecols=use_col)
  df.rename(columns={'NAME': "Name",'STOCKCODE': "Supplier Code","BRAND":'Brand','ONHAND':'Qty'}, inplace = True)
  df = df.loc[df['Brand'].isin(brands_tokeep)]
  df['Qty'] = df['Qty'].astype(str).apply(lambda x:x.replace('+','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())
  df['Qty'] = df['Qty'].astype(int)
  df['Supplier Stock Message'] = "Please contact us to enquire"
  df.loc[df['Qty']>0, 'Supplier Stock Message'] = 'In stock with Supplier'

  df['match'] = df['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  df_supplier_stock = df[['match','Supplier Stock Message']]
  df_p_supplier = df_p_supplier.merge(df_supplier_stock, how='left', on='match').fillna("")

  #Filtered out what needs to be updated

  return df_p_supplier

def kayell():
  df_p_supplier = df_p.loc[df_p['Supplier']=='Kayell']
  df_p_supplier['match'] = df_p_supplier['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  file_path = 'Kayell-'+date.today().strftime("%d-%m-%Y")+'.csv'
  use_col = ['DESCRIPTION', 'STOCK CODE', 'STOCK DESCRIPTION 1', 'AVAILABILITY']

  df = pd.read_csv(file_path, usecols=use_col)
  df.rename(columns={'STOCK DESCRIPTION 1': "Name",'STOCK CODE': "Supplier Code","DESCRIPTION":'Brand','AVAILABILITY':'Qty'}, inplace = True)
  df['Supplier Stock Message'] = "Please contact us to enquire"
  df.loc[df['Qty']=='AVAILABLE', 'Supplier Stock Message'] = 'In stock with Supplier'

  df['match'] = df['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  df_supplier_stock = df[['match','Supplier Stock Message']]
  df_p_supplier = df_p_supplier.merge(df_supplier_stock, how='left', on='match').fillna("")

  #Filtered out what needs to be updated

  return df_p_supplier

def rm():
  df_p_supplier = df_p.loc[df_p['Supplier']=='Rubber Monkey']
  df_p_supplier['match'] = df_p_supplier['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  file_path = 'RM-'+date.today().strftime("%d-%m-%Y")+'.csv'
  use_col = ['Brand', 'Sku', 'Name', 'StockRemaining']

  df = pd.read_csv(file_path, usecols=use_col)
  df.rename(columns={'Name': "Name",'Sku': "Supplier Code","Brand":'Brand','StockRemaining':'Qty'}, inplace = True)
  df['Supplier Stock Message'] = "Please contact us to enquire"
  df.loc[df['Qty']>0, 'Supplier Stock Message'] = 'In stock with Supplier'

  df['match'] = df['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  df_supplier_stock = df[['match','Supplier Stock Message']]
  df_p_supplier = df_p_supplier.merge(df_supplier_stock, how='left', on='match').fillna("")

  #Filtered out what needs to be updated

  return df_p_supplier

def dicker_data():
  df_p_supplier = df_p.loc[df_p['Supplier']=='Dicker Data NZ Limited']
  df_p_supplier['match'] = df_p_supplier['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  file_path = 'DickerData-'+date.today().strftime("%d-%m-%Y")+'.csv'
  use_col = ['Vendor', 'VendorStockCode', 'StockDescription', 'StockAvailable']
  brands_tokeep = ['DJI']
  df = pd.read_csv(file_path, usecols=use_col)
  df.rename(columns={'StockDescription': "Name",'VendorStockCode': "Supplier Code","Vendor":'Brand','StockAvailable':'Qty'}, inplace = True)
  df = df.loc[df['Brand'].isin(brands_tokeep)]

  df['Supplier Stock Message'] = "Please contact us to enquire"
  df.loc[df['Qty']>0, 'Supplier Stock Message'] = 'In stock with Supplier'

  df['match'] = df['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  df_supplier_stock = df[['match','Supplier Stock Message']]
  df_p_supplier = df_p_supplier.merge(df_supplier_stock, how='left', on='match').fillna("")

  #Filtered out what needs to be updated

  return df_p_supplier

def formula():
  df_p_supplier = df_p.loc[df_p['Supplier']=='Formula']
  df_p_supplier['match'] = df_p_supplier['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  file_path = 'Formula-'+date.today().strftime("%d-%m-%Y")+'.csv'
  use_col = ['StockCode', 'StockDescription', 'Brand', 'InStock']
  brands_tokeep = ['JUPIO','PROGRADE']

  df = pd.read_csv(file_path, usecols=use_col)
  df.rename(columns={'StockDescription': "Name",'StockCode': "Supplier Code","Brand":'Brand','InStock':'Qty'}, inplace = True)
  df = df.loc[df['Brand'].isin(brands_tokeep)]
  df['Qty'] = df['Qty'].astype(str).apply(lambda x:x.replace('+','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())
  df['Qty'] = df['Qty'].astype(int)
  df['Supplier Stock Message'] = "Please contact us to enquire"
  df.loc[df['Qty']>0, 'Supplier Stock Message'] = 'In stock with Supplier'

  df['match'] = df['Supplier Code'].astype(str).apply(lambda x:x.replace('$','').replace('\r','').replace(' ','').replace('-','').replace('(','').replace(')','').replace('/','').lower())

  df_supplier_stock = df[['match','Supplier Stock Message']]
  df_p_supplier = df_p_supplier.merge(df_supplier_stock, how='left', on='match').fillna("")

  #Filtered out what needs to be updated

  return df_p_supplier
  
df_update_dove = dove()
df_update_lacklands = lacklands()
df_update_kayell = kayell()
df_update_rm = rm()
df_update_dicker_data = dicker_data()
df_update_formula = formula()

df_SOH = pd.concat([df_update_dove,df_update_lacklands,df_update_kayell,df_update_rm,df_update_dicker_data,df_update_formula])
df_update = df_SOH.loc[df_SOH['Stock Status Out'] != df_SOH['Supplier Stock Message']]

df_SOH.to_csv('Supplier SOH -'+date.today().strftime("%d-%m-%Y")+'.csv')
df_update.to_csv('Supplier Stock Message Update -'+date.today().strftime("%d-%m-%Y")+'.csv')

```
